```
## Allows user to input two arrays that are then used to gather a mean number.

#Input calls for user input 
size = input("Please enter the size of your array: ")

aSize = int(size)
g = []
i = 0

print("Enter the numbers for your first array: ")

## Uses an iterator to continually ask for an input for each space in array g
while i < aSize:
  a1 = input("Enter a number: ")
  a1i = int(a1)
  g.append(a1i)
  i += 1

print(g)  


g2 = []
i2 = 0

print("Enter the floats for your second array: ")

##Same as before for array g2
while i2 < aSize:
  a2 = input("Enter a number: ")
  a2i = float(a2)
  g2.append(a2i)
  i2 += 1
  
print(g2)

##Creates a third array m that multiplies each array position (g and g2) and stores it
m = []

i3 = 0
curIndex = 0
while i3 < aSize:
  pos = g[curIndex]*g2[curIndex]
  m.append(pos)
  i3 += 1
  curIndex += 1
  


print("Mean: Divide by length of array and then add together.")

s = sum(m)
s = s/aSize
print(s)
```
